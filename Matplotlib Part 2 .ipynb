```python
import matplotlib.pyplot as plt
import numpy as np
```

# Pie Chart

```python
a = [34,55,68,27]
mylabel = ['A','B','C','D']
plt.pie(a,labels=mylabel)
```

```python
a = [34,55,68,27]
mylabel = ['A','B','C','D']
plt.pie(a,labels=mylabel,autopct = '%.3f')
```

```python
a = [34,55,68,27]
mylabel = ['A','B','C','D']
plt.pie(a,labels=mylabel,autopct='%2f')
```

```python
a = [34,55,68,27]
mylabel = ['A','B','C','D']
plt.pie(a,labels=mylabel,autopct='%2.2f')
```
